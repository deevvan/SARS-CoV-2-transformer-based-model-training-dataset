```python
import pandas as pd

# Load the input CSV file
input_file = "your/path/RBD_valid_Gen.csv"
df = pd.read_csv(input_file)

# Initialize an empty list to store DataFrames
output_dfs = []

# Group the data by Lineage
grouped = df.groupby("Lineage")

# Iterate through each Lineage group
for lineage, group in grouped:
    generations = group["Generation"].unique()
    generations = sorted([int(gen.replace("Gen", "")) for gen in generations])  # Extract and convert to int

    # Filter for consecutive Generation values
    for i in range(len(generations) - 1):
        gen1, gen2 = generations[i], generations[i + 1]
        if abs(gen2 - gen1) == 1:
            # Get the rows with Generation values gen1 and gen2
            rows_gen1 = group[group["Generation"] == f"Gen{gen1}"]
            rows_gen2 = group[group["Generation"] == f"Gen{gen2}"]

            # Check if there are rows for the specified generations
            if not rows_gen1.empty and not rows_gen2.empty:
                # Create a new DataFrame for the output
                min_len = min(len(rows_gen1), len(rows_gen2))
                new_df = pd.DataFrame({
                    "Lineage": [lineage] * min_len,
                    "GenS1": [f"Gen{gen1}"] * min_len,
                    "GenS2": [f"Gen{gen2}"] * min_len,
                    
                    #"Seq1": rows_gen1["RBD protein"].values[:min_len],
                    #"Seq2": rows_gen2["RBD protein"].values[:min_len],

                    "Seq1": rows_gen1["RBD nucleotide"].values[:min_len],
                    "Seq2": rows_gen2["RBD nucleotide"].values[:min_len],
                    "Variant": rows_gen1["Variant"].values[:min_len],
                })

                # Append the new DataFrame to the list
                output_dfs.append(new_df)

# Concatenate all DataFrames in the list into one output DataFrame
output_df = pd.concat(output_dfs, ignore_index=True)

# Save the output DataFrame to a CSV file
#output_file = "RBD_valid_finetuning_proteins.csv"
output_file = "your/path/RBD_valid_finetuning_nucleotides.csv"
output_df.to_csv(output_file, index=False)
```
