```python
# Python script that will take in csv file that contains RBD nuc sequences and protein sequences to compare 
# Accession ID values with metadata.tsv to access Collection date value for each row 

import pandas as pd
df = pd.read_csv('your/path/filtered_msaCodon_1024_trimmed_RBD_new_1_6mil.csv')

# Load the CSV files
metadata_df = pd.read_csv("your/path/metadata.tsv", sep='\t', dtype=str, low_memory=False, encoding='latin-1')

# Merge the two dataframes on Accession ID
merged_df = df.merge(metadata_df[['Accession ID', 'Collection date']], on='Accession ID', how='left')
```

```python
merged_df.head()
```

```python
# Check and print the number of duplicates in the 'Accession ID' column
num_duplicates = merged_df['Accession ID'].duplicated().sum()
print(f"Number of duplicate Accession IDs: {num_duplicates}")
```

```python
# Python script to divide each Lineage group into 10 Generations unique to each lineage 
# based on Collection date range and select only Lineage groups with atleast 50 sequences


# Convert Collection date to %Y-%m-%d format
merged_df['Collection date'] = pd.to_datetime(merged_df['Collection date'], format='%Y-%m-%d', errors='coerce')
merged_df['Collection date'] = merged_df['Collection date'].dt.strftime('%Y-%m-%d')

# Filter rows with Collection date after November 2019
merged_df['Collection date'] = pd.to_datetime(merged_df['Collection date'])
merged_df = merged_df[merged_df['Collection date'] >= '2019-11-01']

# Group data by Lineage
grouped_data = merged_df.groupby('Lineage')

# Define a function to calculate 'Generation' for each Lineage group
def assign_generation(group):
    date_range = group['Collection date'].max() - group['Collection date'].min()
    bin_size = date_range / 10
    group['Generation'] = pd.cut(group['Collection date'], bins=10, labels=['Gen1', 'Gen2', 'Gen3', 'Gen4', 'Gen5', 'Gen6', 'Gen7', 'Gen8', 'Gen9', 'Gen10'])
    return group


# Filter Lineage groups with more than 10 sequences
filtered_groups = [group for name, group in grouped_data if len(group) > 50]

# Apply the assign_generation function to each filtered Lineage group
processed_data = pd.concat([assign_generation(group) for group in filtered_groups])

# Save the updated data to a new CSV file
output_csv = "your/path/filtered_msaCodon_1024_trimmed_RBD_1_6mil_wt_Gen_50_per_lin.csv"
processed_data.to_csv(output_csv, index=False)

print(f"Output CSV file saved as {output_csv}")
```

```python
# Script to processe CSV containing sequence data to generate comparisons between sequences belonging to
# the same lineage and falling into consecutive generations of lineages dateframe 

# Initialize an empty list to store DataFrames
output_dfs = []

# Group the data by Lineage
grouped = processed_df.groupby("Lineage")

# Iterate through each Lineage group
for lineage, group in grouped:
    generations = group["Generation"].unique()
    generations = sorted([int(gen.replace("Gen", "")) for gen in generations])  # Extract and convert to int

    # Filter for consecutive Generation values
    for i in range(len(generations) - 1):
        gen1, gen2 = generations[i], generations[i + 1]
        if abs(gen2 - gen1) == 1:
            # Get the rows with Generation values gen1 and gen2
            rows_gen1 = group[group["Generation"] == f"Gen{gen1}"]
            rows_gen2 = group[group["Generation"] == f"Gen{gen2}"]

            # Check if there are rows for the specified generations
            if not rows_gen1.empty and not rows_gen2.empty:
                # Create a new DataFrame for the output
                min_len = min(len(rows_gen1), len(rows_gen2))
                new_df = pd.DataFrame({
                    "Lineage": [lineage] * min_len,
                    "GenS1": [f"Gen{gen1}"] * min_len,
                    "GenS2": [f"Gen{gen2}"] * min_len,
                    
                    #"Seq1": rows_gen1["RBD protein"].values[:min_len],
                    #"Seq2": rows_gen2["RBD protein"].values[:min_len],

                    "Seq1": rows_gen1["RBD nucleotide"].values[:min_len],
                    "Seq2": rows_gen2["RBD nucleotide"].values[:min_len],
                    "Variant": rows_gen1["Variant"].values[:min_len],
                })

                # Append the new DataFrame to the list
                output_dfs.append(new_df)

# Concatenate all DataFrames in the list into one output DataFrame
output_df = pd.concat(output_dfs, ignore_index=True)

# Save the output DataFrame to a CSV file
#output_file = "RBD_valid_finetuning_proteins.csv"
output_file = "your/path/filtered_msaCodon_1024_trimmed_RBD_1_6mil_wt_Gen_50_per_lin_finetuning_nucleotides.csv"
output_df.to_csv(output_file, index=False)
```

```python
# Python script that processes above output_file to count the unique occurrences of two columns, 
# Lineage and Variant, and saves the results into a new CSV file. 
import pandas as pd
import os

# Define a function to calculate unique counts of 'Lineage' and 'Variant' columns in the CSV file
def calculate_unique_counts(csv_path, output_path):
    
    # Read the CSV file into a DataFrame
    df = pd.read_csv(csv_path)

    # Check if the required columns ('Lineage' and 'Variant') are in the DataFrame
    if 'Lineage' not in df.columns or 'Variant' not in df.columns:
        # Print an error message if columns are missing and exit the function
        print("Error: The required columns are not present in the input CSV file.")
        return

    # Calculate the count of unique values in the 'Lineage' column
    lineage_counts = df['Lineage'].value_counts().reset_index()
    
    # Rename the columns to 'Lineage' and 'Count' for clarity
    lineage_counts.columns = ['Lineage', 'Count']

    # Calculate the count of unique values in the 'Variant' column
    variant_counts = df['Variant'].value_counts().reset_index()
    
    # Rename the columns to 'Variant' and 'Count' for clarity
    variant_counts.columns = ['Variant', 'Count']

    # Add a new column 'Type' to specify that these counts correspond to 'Lineage'
    lineage_counts['Type'] = 'Lineage'
    
    # Add a new column 'Type' to specify that these counts correspond to 'Variant'
    variant_counts['Type'] = 'Variant'

    # Concatenate lineage and variant counts into a single DataFrame
    combined_counts = pd.concat([lineage_counts, variant_counts], ignore_index=True)

    # Write the concatenated results to the specified output CSV file
    combined_counts.to_csv(output_path, index=False)
    
    # Print a message indicating the output location of the results
    print(f"Results saved to {output_path}")

# Main script execution block
if __name__ == "__main__":
    
    # Define the main directory path where input and output files are located
    main_dir = "your/path/"
    
    # Define the full path to the input CSV file containing 'Lineage' and 'Variant' columns
    csv_path = os.path.join(main_dir, "filtered_msaCodon_1024_trimmed_RBD_1_6mil_wt_Gen_50_per_lin.csv")  # Update this path as needed
    
    # Define the full path to the output CSV file for saving the results
    output_path = os.path.join(main_dir, "lineage_variant_counts_1_6_mil.csv")
    
    # Call the function to calculate unique counts and save results to the output file
    calculate_unique_counts(csv_path, output_path)
```
